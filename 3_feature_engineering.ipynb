```python
import pandas as pd
import numpy as np
from functions import *
from parameters import *
```

```python
df = pd.read_csv(path_processed_data + '/nba_all_stats.csv')
```

1- Versatility to play in different positions

```python
df['Versatility Score'] = round(df[['PG%', 'SG%', 'SF%', 'PF%', 'C%']].apply(shannon_entropy, axis=1),4)
df[['Player','PG%', 'SG%', 'SF%', 'PF%', 'C%', 'Versatility Score']].head(10)
```

2- Blocks for each foul committed on a shooter

```python
df['Blocks/Foul Ratio'] = round(df['BLK'] / (df['Fouls Commited Shoot']+epsilon),4)
df[['Player', 'BLK', 'Fouls Commited Shoot', 'Blocks/Foul Ratio']].head(10)
```

3- Foul and score for each foul received while shooting

```python
df['And1 Efectiveness'] = round(df['And1'] / df['Fouls Drawn Shoot'],4)
df['And1 Efectiveness'].fillna(0, inplace=True)
df[['Player', 'And1', 'Fouls Drawn Shoot', 'And1 Efectiveness']].head(10)
```

4- Assists per lost ball due to bad pass

```python
df['AST/TOV Ratio'] = df['AST'] / (df['BadPass'])
df.loc[df['AST/TOV Ratio'] == np.inf, 'AST/TOV Ratio'] = 0
df[['Player', 'AST', 'BadPass', 'AST/TOV Ratio']].head(10)
```

5- Offensive rebounds per defensive rebound

```python
df['ORB/DRB'] = round(df['ORB'] / df['DRB'],4)
df[['Player', 'ORB', 'DRB', 'ORB/DRB']].head(10)
```

6- Net plus-minus: Box +/- offenive - Box +/- defensive

```python
df['Net Plus Minus'] = round(df['OBPM'] - df['DBPM'],4)
df[['Player', 'OBPM', 'DBPM', 'Net Plus Minus']].head(10)
```

7- Percentage of 3-point shots attempted

```python
df['3P Shots Attempted %'] = round(df['3PA'] / (df['3PA'] + df['2PA']),4)
df[['Player', '3PA', '2PA', '3P Shots Attempted %']].head(10)
```

8- Percentage of 2-point shots attempted

```python
df['2P Shots Attempted %'] = round(df['2PA'] / (df['3PA'] + df['2PA']),4)
df[['Player', '3PA', '2PA', '2P Shots Attempted %']].head(10)
```

9- Personal fouls efficiency

```python
df['PF Efficiency'] = round((df['STL'] + df['BLK']) / df['PF'],4)
df[['Player', 'STL', 'BLK', 'PF','PF Efficiency']].head(10)
```

10- Points per shot attempt

```python
df['Points per Shot attempted'] = round((df['2P']*2 + df['3P']*3) / df['FGA'],4)
df[['Player', '2P', '3P','FGA','Points per Shot attempted']].head(10)
```

11- Games starting percentage

```python
df['Starting Percentage'] = round(df['GS'] / df['G'],4)
df[['Player', 'GS', 'G','Starting Percentage']].head(10)
```
